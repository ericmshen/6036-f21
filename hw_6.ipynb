# **Softmax Implementation**

You may find it helpful to implement a softmax function to use for solving some of the problems, or work it out by hand. You can then call your softmax on z, which you can also find computationally or by hand. 

```
# Implement Softmax
def SM(z):
  return np.exp(z)/np.sum(np.exp(z))
```
